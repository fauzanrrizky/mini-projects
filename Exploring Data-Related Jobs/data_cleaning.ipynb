```python
def unzip_file(zip_file_path, extract_dir):
    """
    Unzips a zip file to a specified extraction directory.
    
    Parameters:
    zip_file_path (str): The path to the zip file.
    extract_dir (str): The directory where the contents should be extracted.
    
    Returns:
    None
    """
    # Create the extraction directory if it doesn't exist
    if not os.path.exists(extract_dir):
        os.makedirs(extract_dir)
    
    # Unzip the file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print(f"Extracted {zip_file_path} to {extract_dir}")
```

## Exploring Dataset

**Missing Values**

```python
def missing_val(df):
    # Calculate the number of missing values for each column
    missing_values = df.isnull().sum()

    # Calculate the total number of rows
    total_rows = len(df)

    # Calculate the percentage of missing values for each column
    missing_percentage = (missing_values / total_rows) * 100

    # Create a DataFrame to display the results
    missing_percentage_df = pd.DataFrame({'Column': df.columns, 'Missing Percentage': missing_percentage})

    # Display the missing percentage for each column
    print(missing_percentage_df)
```

## Data Cleaning & Manipulation

**Remove Unwanted Columns**
columns that will not bring insights for data analysis. or too much missing values

```python
# List columns to remove: df1
columns_to_remove_df1 = ["Unnamed: 0","index","description","commute_time","salary_avg","salary_min","salary_max","salary_hourly","salary_standardized","salary_rate","extensions","job_id","thumbnail",
                     "salary_pay","salary_yearly","salary","search_term","posted_at","search_location"]

# Drop columns: df1
df1 = df1.drop(columns=columns_to_remove_df1)

# Drop columns: df2
df2 = df2.drop(columns=['job_summary'])
df2 = df2.drop(columns=['job_link'])
df2 = df2.drop(columns=['search_city'])
df2 = df2.drop(columns=['search_country'])
```

**Remove redundant prefix on df1**

```python
df1['via'] = df1['via'].str.replace('via','')
```

**Add new column post_time on df1**

Dropping timestamp from date_time, and make a new column 'post_time' so the column and data type would match df2

```python
# Ensure the date_time column is in datetime format
df1['date_time'] = pd.to_datetime(df1['date_time'])

# Extract only the date part
df1['post_time'] = df1['date_time'].dt.date

#drop date_time
df1=df1.drop(columns=['date_time'])
```

**Change column names on both dataframes for merging dataframe**

```python
new_column_names_df2 = {
    'job_title': 'title',
    'company': 'company_name',
    'job_location': 'location',
    'first_seen' : 'post_time',
    'job_location' : 'location',
}

new_column_names_df1 = {
    'description_tokens': 'job_skills',
    'work_from_home': 'job_type',
}
```

```python
# Rename the columns
df2.rename(columns=new_column_names_df2, inplace=True)
df1.rename(columns=new_column_names_df1, inplace=True)
```

**Imputing value in 'job_type'**

Imputing 'null' value in column "job_type" to 'Onsite' (Basic work scenario), and 'True' value to Hybrid/Remote

Ignore df2 because there aren't any null values in the 'job_type' column

```python
# Replace null values in 'job_type' column with 'Onsite'
df1['job_type'] = df1['job_type'].fillna('Onsite')

# Replace True values in 'job_type' column with 'Hybrid'
df1['job_type'] = df1['job_type'].replace(True, 'Hybrid')
```

**Make new column in df1 to map job level**

```python
# Get job level from schedule_type
df1['job level'] = df1['schedule_type'].apply(lambda x: 'Internship' if pd.notna(x) and 'Internship & Entry Level' in x else 'Non Intern')

#Drop schedule_type
df1 = df1.drop(columns=['schedule_type'])
```

```python
# Determining job level
def determine_job_level(title):
    title_lower = title.lower() if pd.notna(title) else ''
    if any(keyword in title_lower for keyword in ['internship', 'intern', 'entry']):
        return 'Internship & Entry Level'
    elif 'associate' in title_lower:
        return 'Associate'
    elif 'director' in title_lower:
        return 'Director & Executives'
    else:
        return 'Mid Senior'

df1['job level'] = df1['title'].apply(determine_job_level)
```

```python
# Change job_level value in df2 to match df1
df2['job level'] = df2['job level'].replace('Mid senior', 'Mid Senior')
```

**Adjust job_skills column on df1 to match df2**

```python
def clean_skill_string(skill_str):
    # Hapus tanda kurung siku dan tanda kutip
    cleaned_str = re.sub(r"[\[\]']", "", skill_str)
    # Pisahkan kata-kata dengan koma dan spasi
    return ', '.join(word.strip().capitalize() for word in cleaned_str.split(','))

df1['job_skills'] = df1['job_skills'].apply(clean_skill_string)

# Change empty string to NaN
df1['job_skills'].replace("", np.nan, inplace=True)
```

## Concat dataframe

```python
# Add 'via' column to df2 and fill with NaN
df2['via'] = np.nan

# Ensure columns are in the same order
df2 = df2[df1.columns]

# Concatenate DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)
```

**Make new column: title_category**
- Data Analyst
- Business Intelligence
- Data Scientist
- Data Engineer

```python
# Determine title category
def categorize_title(title):
    title = title.lower()
    if any(keyword in title for keyword in ['intelligence', 'business','bi']):
        return 'Business Intelligence'
    elif any(keyword in title for keyword in ['science', 'scientist', 'machine','learning']):
        return 'Data Scientist'
    elif any(keyword in title for keyword in ['engineer', 'architect']):
        return 'Data Engineer'
    elif any(keyword in title for keyword in ['data analyst']):
        return 'Data Analyst'
    else:
        return 'Other'

combined_df['title_category'] = combined_df['title'].apply(categorize_title)
```

**Make skill columns: for visualization purpose**

```python
# Separate jobs to list
combined_df['job_skills'] = combined_df['job_skills'].str.split(', ')
```

```python
all_skills = []
# Add all skills to list
for skills in combined_df['job_skills'].dropna():
    all_skills.extend([skill.lower() for skill in skills])

# Hitung frekuensi setiap skill
skill_counts = Counter(all_skills)

# Cetak total dan top 100 skills
print("Total Skills:", len(skill_counts))
```

Because the total skill is so large (so variative), I will just get top 50 most frequent skills requirements and make new columns based on that

```python
# get top 50 most frequent skills
top_50_skills_count = skill_counts.most_common(50)

print("Top 50 Skills:")
for skill, count in top_50_skills_count:
    print(f"{skill}: {count}")
```

```python
# Make new column for every top 50 skill, with a value of 1 or 0
top_50_skills = [skill for skill, count in skill_counts.most_common(50)]
for skill in top_50_skills:
    col_name = f"skill_{skill}"
    combined_df[col_name] = combined_df['job_skills'].apply(
        lambda x: 1 if isinstance(x, list) and skill in [s.lower() for s in x] else 0
    )

# Show result
combined_df.head()
```
